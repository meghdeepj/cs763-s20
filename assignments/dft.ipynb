### Histograms Open-CV

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('../data/spoons.jpeg', 0)
```

```python
cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
f=np.fft.fft2(img)
fshift=np.fft.fftshift(f)
```

```python
mag_spec=20*np.log(np.abs(fshift))
```

```python
res=np.hstack((img,mag_spec))
```

```python
cv2.imshow('imgfft', res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.subplot(121),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(mag_spec, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
rows,cols=img.shape
crow,ccol=rows//2,cols//2
fshift[crow-15:crow+16, ccol-15:ccol+16]=0
f_i=np.fft.ifftshift(fshift)
lp_img=np.fft.ifft2(f_i)
lp_img=np.real(lp_img)
```

```python
plt.subplot(121),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(lp_img, cmap = 'gray')
plt.title('HFTd image'), plt.xticks([]), plt.yticks([])
plt.show()
```
